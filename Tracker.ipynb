```python
# ====================================================================
# SCRIPT 2: TAGUIG RIVER POLLUTION DATA TRACKER
# ====================================================================

# Number of plastic waste items collected per testing station
river_trash_count = [142, 89, 210, 55, 320, 115]

# Using built-in functions to extract mathematical data
total_trash = sum(river_trash_count)
total_stations = len(river_trash_count)
average_trash = total_trash / total_stations

# Outputting a formatted field research report
print("--- TAGUIG RIVER ENVIRONMENTAL REPORT ---")
print(f"Total river stations monitored: {total_stations}")
print(f"Total plastic waste recovered: {total_trash} items")
print(f"Average trash payload per station: {average_trash:.2f} items")
print("-----------------------------------------")
```
